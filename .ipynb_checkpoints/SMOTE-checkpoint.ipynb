# SMOTE 

*SMOTE: Synthetic Minority Oversampling Technique*

Smote increases the minority class examples by 
1. Select a random data point say 'a'
2. Find nearest neighbors for 'a' (typically k=5) 
3. Pick one of the nearest neighbor at random say 'b'
4. Finally, build a synthetic data point of minority class as a convex combination of 'a' & 'b'

Repeat above steps to create representative population of minority class


* ** SMOTE works well in combination of under sampling of Majority class *

> - __Upside__:   SMOTE works to cause the classifier to build larger decision regions that contain nearby minority class points.
> -__Downsides__: synthetic observations are created without considering the majority class, possibly resulting in ambiguous observations if there is a strong overlap for the classes. Designed only for Tabular data (i.e doesn't work on data such as images)


#### Helper functions

```python
def plot_data(X,y):
    #check class distribution
    counter = Counter(y)
    print(counter)
    # scatter plot of examples by class label
    for label, _ in counter.items():
        row_ix = np.where(y == label)[0]
        pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
    pyplot.legend()
    pyplot.show()
    return
```

#### Import libraries

```python
#ML libraries
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE, BorderlineSMOTE, SVMSMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.tree import DecisionTreeClassifier


#Math libraries
import numpy as np

#Helper libraries 
from collections import Counter
from numpy import mean

#Visualization libraries
from matplotlib import pyplot
```

```python
#create Dummy data set
X, y = make_classification(n_samples=10000, n_features=2, n_redundant=0,
	n_clusters_per_class=1, weights=[0.99], flip_y=0, random_state=1)

#Check class distribution & plot data
plot_data(X,y)
```

```python
#Oversample minority class with SMOTE 
minority_oversample=SMOTE(sampling_strategy=0.1)

# transform the dataset
X_smote_0, y_smote_0 = minority_oversample.fit_resample(X, y)


#Check class distribution & plot data
plot_data(X_smote,y_smote)
```

```python
#Oversample minority class with SMOTE & undersample majority class
minority_oversample=SMOTE(sampling_strategy=0.1)
majority_undersample=RandomUnderSampler(sampling_strategy=0.5)
#Create a pipeline
steps = [('o', minority_oversample), ('u', majority_undersample)]
pipeline = Pipeline(steps=steps)
# transform the dataset
X_smote, y_smote = pipeline.fit_resample(X, y)


#Check class distribution & plot data
plot_data(X_smote,y_smote)
```

## 1.0 Classification Example with SMOTE

```python
#Classification on Imbalanced dataset 

# define model
model = DecisionTreeClassifier()
# evaluate pipeline
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(model, X, y, scoring='roc_auc', cv=cv, n_jobs=-1)
print('Mean ROC AUC: %.3f' % mean(scores))
```

```python
#Classifcation on resampled data: Only Upsample minority with SMOTE

# define model
model = DecisionTreeClassifier()
# evaluate pipeline
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(model, X_smote_0, y_smote_0, scoring='roc_auc', cv=cv, n_jobs=-1)
print('Mean ROC AUC: %.3f' % mean(scores))
```

```python
#Classifcation on resampled data: Upsample minority with SMOTE & Downsample majority

# define model
model = DecisionTreeClassifier()
# evaluate pipeline
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(model, X_smote, y_smote, scoring='roc_auc', cv=cv, n_jobs=-1)
print('Mean ROC AUC: %.3f' % mean(scores))
```

## 1.1 SMOTE: Find best K parameter

Another area to explore would be to test different values of the k-nearest neighbors selected in the SMOTE procedure when each new synthetic example is created. The default is k=5, although larger or smaller values will influence the types of examples created, and in turn, may impact the performance of the model.

```python
k_values = [1, 2, 3, 4, 5, 6, 7]
for k in k_values:
	# define pipeline
    model = DecisionTreeClassifier()
    minority_oversample=SMOTE(sampling_strategy=0.1)
    majority_undersample=RandomUnderSampler(sampling_strategy=0.5)
    steps = [('o', minority_oversample), ('u', majority_undersample), ('m', model)]
    pipeline = Pipeline(steps=steps)
	# evaluate pipeline
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    scores = cross_val_score(pipeline, X, y, scoring='roc_auc', cv=cv, n_jobs=-1)
    score = mean(scores)
    print('> k=%d, Mean ROC AUC: %.3f' % (k, score))
```

## 2. SMOTE with Selective Synthetic Sample Generation

#### 2.1 Borderline-SMOTE

Select only the synthetic instances of the minority class observation that are misclassified, <br />
thus providing more resolution where it is required in the region of edge/border of decision boundary <br />
- KNN classifier is used to decide the decision boundary

```python
oversample = BorderlineSMOTE()
X_borderline, y_borderline = oversample.fit_resample(X, y)
plot_data(X_borderline,y_borderline)
```

#### 2.2 Borderline SMOTE SVM

An SVM is used to locate the decision boundary defined by the support vectors <br /> 
and examples in the minority class that close to the support vectors <br /> 
become the focus for generating synthetic examples
- Replaces KNN classfier with SVM to determine the decision Boundary

```python
oversample = SVMSMOTE()
X_borderline_svm, y_borderline_svm = oversample.fit_resample(X, y)
plot_data(X_borderline_svm,y_borderline_svm)
```

#### 2.3 Adaptive Synthetic Sampling (ADASYN)

ADASYN is based on the idea of adaptively generating minority data samples according to their distributions: more synthetic data is generated for minority class samples that are harder to learn compared to those minority samples that are easier to learn.

- Generate synthetic samples in regions where the density of minority class is low, and fewer or none where density is high
- i.e generating synthetic samples inversely proportional to the density of the observations in the minority class

Ref: https://ieeexplore.ieee.org/document/4633969

```python
oversample = ADASYN()
X_adasyn, y_adasyn = oversample.fit_resample(X, y)
plot_data(X_adasyn,y_adasyn)
```
